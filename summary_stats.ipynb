```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('cleaned_data.csv', index_col=0)
developer_data = pd.read_csv('developer_data.csv', index_col=0)
new_app_data = pd.read_csv('new_app_data.csv', index_col=0)
```

```python
print(len(data))
print(len(developer_data))
print(len(new_app_data))
```

```python
# App summary stats: Basic

# basic
basic_v0 = data[['age', 'ratings_score', 'ratings_count', 'price', 'purchases']].describe().transpose()
basic = basic_v0[['min', '50%', 'mean', 'max', 'count']]
basic
```

```python
# App summary stats: Privacy

# privacy
privacy_v0 = data[['data_used_to_track_you', 'data_linked_to_you', 'data_not_linked_to_you', 'data_not_collected', 'no_details_provided']].describe().transpose()

privacy = privacy_v0[['min', '50%', 'mean', 'max', 'count']]
privacy
```

```python
# App summary stats: Category
cat_v0 = data[['books', 'business', 'developer_tools', 'education', 'entertainment', 'finance', 'food_&_drink', 'games', 'graphics_&_design', 'health_&_fitness', 'lifestyle', 'magazines_&_newspapers', 'medical', 'music', 'navigation', 'news', 'photo_&_video', 'productivity', 'reference', 'shopping', 'social_networking', 'sports', 'stickers', 'travel', 'utilities', 'weather']].describe().transpose()

cat = cat_v0[['min', '50%', 'mean', 'max', 'count']]
cat
```

```python
# Tracker apps stats

trackers = data[data.data_used_to_track_you == 1]
print(len(trackers))
track_v0 = trackers[['age', 'ratings_score', 'ratings_count', 'price', 'purchases']].describe().transpose()
track = track_v0[['min', '50%', 'mean', 'max', 'count']]
track
```

```python
# Developer summary data

dev_v0 = developer_data[['dev_total_apps', 'dev_total_ratings_count', 'dev_avg_ratings_count', 'dev_avg_ratings_score', 'dev_category_count', 'dev_new_apps',]].describe().transpose()

dev = dev_v0[['min', '50%', 'mean', 'max', 'count']]
dev
```

```python
# New app summary stats: Basic
new_app_v0 = new_app_data[['age', 'ratings_score', 'ratings_count', 'price', 'purchases']].describe().transpose()

new_app = new_app_v0[['min', '50%', 'mean', 'max', 'count']]
new_app
```

```python
# New app summary stats: Privacy
new_app_p_v0 = new_app_data[['data_used_to_track_you', 'data_linked_to_you', 'data_not_linked_to_you', 'data_not_collected', 'no_details_provided']].describe().transpose()

new_app_p = new_app_p_v0[['min', '50%', 'mean', 'max', 'count']]
new_app_p
```

```python
# New app summary stats: Category
new_app_c_v0 = new_app_data[['books', 'business', 'developer_tools', 'education', 'entertainment', 'finance', 'food_&_drink', 'games', 'graphics_&_design', 'health_&_fitness', 'lifestyle', 'magazines_&_newspapers', 'medical', 'music', 'navigation', 'news', 'photo_&_video', 'productivity', 'reference', 'shopping', 'social_networking', 'sports', 'stickers', 'travel', 'utilities', 'weather']].describe().transpose()

new_app_c = new_app_c_v0[['min', '50%', 'mean', 'max', 'count']]
new_app_c
```
